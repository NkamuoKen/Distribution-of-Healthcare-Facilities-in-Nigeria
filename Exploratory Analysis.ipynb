```python
# Import pandas

import pandas as pd
```

```python
data = pd.read_csv('cleaned_hospital&clinic_data.csv')
```

```python
# Get a copy of our csv file

df = data.copy()
df
```

```python
# We want to understand the columns of our dataset

df.info()
```

```python
# Let us check the general descriptive statistics of all the dataset

pd.set_option('display.max_columns', None)
df.describe(include='all')
```

```python
# Let us remove some unnecessary inconsistency from our latitude column

df['latitude'] = df['latitude'].replace('5.623921`66666667', '5.623921')
df['latitude'] = df['latitude'].replace(':5.07063007', '5.07063007')
```

```python
column_name = 'latitude'
df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
df = df.dropna(subset=[column_name])
df = df.reset_index(drop=True)
```

```python
# So now i can now convert it to a float dtype unlike when it had the inconsistencies

df['latitude'] = df['latitude'].astype(float)
```

```python
# Explore our categorical features

categorical_columns = ['state', 'ownership', 'facility_level', 'registration_status', 'license_status', 'inpatient', 
                       'outpatient', 'onsite_laboratory', 'facility_level_option', 'lga', 'ward', 'ambulance_services', 
                       'operational_days', 'physical_location', 'ownership_type']

for col in categorical_columns:
    print(f"\n{col} Distribution:")
    print(df[col].value_counts())
    print("\n")
```

```python
# Explore our numeric our features
numeric_columns = ['longitude', 'latitude', 'operational_hours', 'doctors', 'pharmacists', 'dentist', 'pharmacy_technicians', 
                   'nurses', 'lab_scientists', 'midwifes', 'lab_technicians', 'nurse_midwife', 'him_officers', 
                   'community_health_officer', 'community_extension_workers', 'jun_community_extension_worker', 
                   'dental_technicians', 'env_health_officers', 'beds', 'attendants']

for col in numeric_columns:
    print(f"\n{col} Summary Statistics:")
    print(df[col].describe())
    print("\n")
```

```python
correlation_matrix = df[numeric_columns].corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

From the correlation matrix we can see that there is a strong positive correlation between pharmacists and doctors which is 0.81 now let me explain, when the numbers in the box are closer to one then there is a strong positive correlation but when it is closer to -1 then there is a strong negative correlation and when it is zero it means no correlation. So since the pharmacists have a strong correlation with doctors it means that the more doctors in a Health Facility increases the more the pharmacists will also increase, this also go to some other Health Professionals which is doctors to lab_scientists, doctors to nurse_midwife, pharmacists to lab_scientists, and lab_scientists to nurse_midwife. For strong negative correlation we have only between latitude and operational_hours which is not that important for our analysis. Now the 1.00 that appeared diagonally is just means a column correlated to itself which will obviously be perfect.

```python
# Let me turn some columns from float to int, because it doesn't make sense as floats. For 2.5 doctors doesn't make any sense so
# it better that are in int format

numerical_columns = ['doctors', 'pharmacists', 'dentist', 'pharmacy_technicians', 'nurses', 'lab_scientists', 'midwifes', 
                     'lab_technicians', 'nurse_midwife', 'him_officers', 'community_health_officer', 
                     'community_extension_workers', 'jun_community_extension_worker', 'dental_technicians', 
                     'env_health_officers', 'beds', 'attendants']

df[numerical_columns] = df[numerical_columns].round(0).astype(int)
```
